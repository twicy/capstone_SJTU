```python
import pandas as pd
import random
```

```python
df_cons=pd.read_csv('../data/rawData/consumption.csv')
df_prod=pd.read_csv('../data/rawData/production.csv')
df_warning=pd.read_csv('../data/rawData/warning.csv')
df_workshift=pd.read_csv=('../data/rawData/workShift.csv')
```

```python
label_set_cons=pd.unique(df_cons['lable_CHinese'])
label_Ch_En_cons={}
for k, v in df_cons.iloc[:,[2,3]].values:
    label_Ch_En_cons[k]=v

label_set_prod=pd.unique(df_prod['lable_CHinese'])
label_Ch_En_prod={}
for k, v in df_prod.iloc[:,[2,3]].values:
    label_Ch_En_prod[k]=v
    
```

```python

df_all_consumptions=pd.DataFrame()
for k, v in label_Ch_En_cons.items():
    df_all_consumptions=df_all_consumptions.append({'Chinese_label':k,'English_label':v},ignore_index=True)
df_all_consumptions.reset_index(inplace=True)
    
df_all_productions=pd.DataFrame()
for k, v in label_Ch_En_prod.items():
    df_all_productions=df_all_productions.append({'Chinese_label':k,'English_label':v},ignore_index=True)
df_all_productions.reset_index(inplace=True)
```

```python
df_all_warnings=pd.DataFrame()
for idx,row in df_warning.iterrows():
    df_all_warnings=df_all_warnings.append({'index':idx,'label_Chinese':row['lable_CHinese'],'label_English':row['lable_English'],\
                            'function_name':row['funCTion_name'],'function_Chinese':row['funCTion_label_p'],'function_English':row['funCTion_label_s'],\
                           'group_label_Chinese':row['group_label_p'],'group_label_English':row['group_label_s'],'machine_obj_Chinese':row['maCHine_label_p'],'machine_obj_English':row['maCHine_obj']},ignore_index=True)

df_all_warnings
```

```python
week_shift_consumption=pd.DataFrame()
week_shift_production=pd.DataFrame()
weeks=[week for week in range(1,100)]
shifts=[1,2,3]
for week in weeks:
    for shift in shifts:
        for index in df_all_consumptions['index']:
            value=random.randint(0,100)
            week_shift_consumption=week_shift_consumption.append({'week':week,'shift':shift,'consumption_id':index,'consumption':value},ignore_index=True)
week_shift_consumption
```

```python

for week in weeks:
    for shift in shifts:
        for index in df_all_productions['index']:
            value=random.randint(0,100)
            week_shift_production=week_shift_production.append({'week':week,'shift':shift,'production_id':index,'production':value},ignore_index=True)
week_shift_consumption
```

```python
week_shift_warning=pd.DataFrame()
for week in weeks:
    for shift in shifts:
        for index in df_all_warnings['index']:
            week_shift_warning=week_shift_warning.append({'week':week,'shift':shift,'warning_id':index,'occur':random.choice([True,False])},ignore_index=True)
week_shift_warning
```

```python
df_all_consumptions.rename(columns={'Chinese_label':'label_Chinese','English_label':'label_English'},inplace=True)
df_all_productions.rename(columns={'Chinese_label':'label_Chinese','English_label':'label_English'},inplace=True)
```

```python
week_shift_consumption.to_csv('../data/tables/week_shift_consumption.csv')
week_shift_production.to_csv('../data/tables/week_shift_production.csv')
week_shift_warning.to_csv('../data/tables/week_shift_warning.csv')
df_all_consumptions.to_csv('../data/tables/consumption_list.csv')
df_all_productions.to_csv('../data/tables/production_list.csv')
df_all_warnings.to_csv('../data/tables/warning_list.csv')
```
